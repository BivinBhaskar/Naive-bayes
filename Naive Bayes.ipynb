```python
# feature engineering
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
train1=train.iloc[:,0:13]
train1
```

```python
train1=train1.apply(le.fit_transform)
train1
```

```python
test1=test.iloc[:,0:13]
test1
```

```python
test1=test1.apply(le.fit_transform)
test1
```

```python
finaltrain=pd.concat([train1,train['Salary']],axis=1)
finaltrain
```

```python
finaltest=pd.concat([test1,test['Salary']],axis=1)
finaltest
```

```python
## train test split
X=finaltrain.iloc[:,0:13]
X
```

```python
Y=finaltrain[['Salary']]
Y
```

```python
x=finaltest.iloc[:,0:13]
y=finaltest[['Salary']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,train_size=0.75,random_state=20)
```

### Multinomial bayes model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
mb=MultinomialNB()
```

```python
mb.fit(xtrain,ytrain)
```

```python
trainpredict=pd.DataFrame(mb.predict(X))
trainpredict
```

```python
# score for train data
mb.score(X,Y)
```

```python
# testing using test data
testpred=pd.DataFrame(mb.predict(x))
testpred
```

```python
# score for test data
mb.score(x,y)
```
